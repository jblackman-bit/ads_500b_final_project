Ideas
    - Multiple Linear Regression (MLR) - predict price (dependent variable) from some combination of some of the independent variables. "What is the specific dollar value of this house?"
    - Clustering (K-means clustering) - Grouping geographic regions based on a number of criteria. "What natural 'neighborhood types' exist here?"
        - example https://www.youtube.com/watch?v=Se28XHI2_xE
    - Classification (random forest or logistic regression) - Categorize a house based on a feature (binary target, i.e. price > $1M). "Is this house 'Luxury'?"
    - Density Estimation (Kernel Density Estimation) - Where does the most value for money exist as a region rather than individual houses? "Where is the highest concentration of $ value in the county?"

Steps for MLR
Data pre-processing/cleaning (EDA process)
    - Handle missing values and outliers
    - Engineer any features (year built minus date after converting date to year, zip code regions through first three numbers, etc.)
    - Scaling variables or any other data transformations for non-linear relationships, also interaction effects. This may change after data visualization.
Data Visualization
    - Correlation heatmap, box plots (both individual and over key groups), scatter plots, distribution plots, qq plots.
Data Analysis
    - Check for multicollinearity (VIF), drop variables based on overlap from correlation heatmaps to reduce VIF.
    - Check for model fit - do not over fit.
    - Check residuals (heteroskedasticity, Durbin-Watson)
    - Repeat until satisfied.

Steps for Clustering
Data pre-processing/cleaning (EDA process)
    - Select feature for clustering.
    - Handle missing values and outliers
    - Engineer any features (year built minus date after converting date to year, zip code regions through first three numbers, etc.)
    - Scaling variables since clustering algorithms are based off of the euclidean distance.
    - Drop any variables that do not help this analysis (ID, date, etc.)
Data Visualization
    - geospatial plots (if lat and long or zip code is used), box plots, pair plots.
Data Analysis
    - Elbow method, silhouette score, cluster profiling (where we define groups like luxury, waterfront, entry-level, investment property)

Steps for Classification
    - Data pre-processing/cleaning (EDA process)
    - Define Target Label: Create a categorical target variable (e.g., "Luxury" vs "Standard" based on price percentiles or grade).
    - Handle Class Imbalance: Check if one category is underrepresented; apply techniques like SMOTE or stratified sampling if necessary.
    - Feature Selection: Identify features with high predictive power for the specific categories (e.g., view, waterfront, grade).
    - Encoding: Convert any remaining categorical strings into numerical format (One-Hot or Label Encoding).
Data Visualization
    - Count Plots: Visualize the distribution of the target classes to check for balance.
    - Facet Grids: Plot distributions of key features (like sqft_living) grouped by the target class to see separation.
    - Confusion Matrix Heatmap: Visualize the accuracy of predictions across all categories.
    - ROC/AUC Curves: Plot the true positive rate against the false positive rate to evaluate model thresholding.
Data Analysis
    - Train/Test Split: Partition data to ensure the model is evaluated on unseen data.
    - Model Comparison: Test multiple algorithms (e.g., Random Forest, Logistic Regression, XGBoost) to find the best fit.
    - Feature Importance: Extract and rank which variables (like location or sqft) were the primary drivers of the classification.
    - Evaluation Metrics: Calculate Precision, Recall, and F1-Score to ensure the model isn't just "guessing" the majority class.

Steps for Density Estimation
    - Data pre-processing/cleaning (EDA process)
    - Select Spatial/Value Features: Focus on lat, long, or price to define the "surface" you are measuring.
    - Handle Spatial Outliers: Remove or clip extreme geographic coordinates that may distort the density "cloud."
    - Feature Scaling: Normalize variables if performing multi-dimensional density estimation (e.g., price density over a geographic area).
    - Log Transformations: Apply to skewed variables like price to ensure the density estimation isn't dominated by a few extreme values.
Data Visualization
    - KDE Plots (2D): Create contour or "shaded" maps to show where properties are most concentrated.
    - Joint Plots: Use scatter plots with marginal histograms to see density along individual axes simultaneously.
    - Hexbin Plots: Use hexagonal binning for a cleaner look at high-density areas compared to standard scatter plots.
    - Heatmaps: Overlay a density map onto a geographic representation of the county.
Data Analysis
    - Bandwidth Selection: Adjust the smoothing parameter (bandwidth) to find the balance between "noisy" and "over-smoothed" data.
    - Modality Check: Analyze if the market has a single "hot center" or multiple distinct sub-market peaks.
    - Thresholding: Define "High-Density" zones vs. "Sparse" zones to identify underserved or over-saturated markets.
    - Overlap Analysis: Compare the density of one feature (e.g., grade) against another (e.g., price) to find geographic anomalies.
